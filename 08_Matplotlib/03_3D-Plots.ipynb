### Scatter plot 3d

```python
fig=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(df['runs'],df['avg'],df['strike_rate'])
```

### Line Plot

```python
x=[1,2,3,4,25]
y=[4,5,6,7,0]
z=[8,9,10,11,7]
```

```python
fig=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(x,y,z)
ax.plot3D(x,y,z,color='red')
```

### surface plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx,yy=np.meshgrid(x,y)
zz=xx**2+yy**2
```

```python
xx.shape
```

```python
zz.shape
```

```python
fig=plt.figure(figsize=(15,7))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,zz,cmap='viridis')
fig.colorbar(p)
```

### contour 

```python
fig=plt.figure(figsize=(15,7))
ax=plt.subplot()
p=ax.contour(xx,yy,zz,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(15,7))
ax=plt.subplot()
p=ax.contourf(xx,yy,zz,cmap='viridis')
fig.colorbar(p)
```
