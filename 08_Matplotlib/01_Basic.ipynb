```python
# Datavisulisation
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.plot(data['index'],data['V Kohli'])
plt.plot(data['index'],data['RG Sharma'])
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(data['index'],data['V Kohli'],color="red")
plt.plot(data['index'],data['RG Sharma'],color="black")
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(data['index'],data['V Kohli'],color="red",linestyle='dotted')
plt.plot(data['index'],data['RG Sharma'],color="black")
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(data['index'],data['V Kohli'],color="red",linestyle='dashdot')
plt.plot(data['index'],data['RG Sharma'],color="black",linewidth=5)
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(data['index'],data['V Kohli'],color="red",linestyle='dashdot',marker='D',markersize=10)
plt.plot(data['index'],data['RG Sharma'],color="black",linewidth=5)
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(data['index'],data['V Kohli'],color="red",linestyle='dashdot',label='Virat')
plt.plot(data['index'],data['RG Sharma'],color="black",linewidth=2,label="Rohit")
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
# plt.legend()
plt.legend(loc="upper left")
```

```python
plt.plot(data['index'],data['V Kohli'],color="red",linestyle='dashdot',label='Virath')
plt.plot(data['index'],data['RG Sharma'],color="black",linewidth=2,label="Rohit")
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
plt.legend()
plt.xlim(2010,2014)
plt.ylim(200,600)
plt.legend(loc="lower right")
```

```python
# Grid
plt.plot(data['index'],data['V Kohli'],color="red",linestyle='dashdot',label='Virath')
plt.plot(data['index'],data['RG Sharma'],color="black",linewidth=2,label="Rohit",marker='s')
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
plt.legend()
plt.legend(loc="upper left")
plt.grid()
```

```python
plt.plot(data['index'],data['V Kohli'],color="red",linestyle='dashdot',label='Virath')
plt.plot(data['index'],data['RG Sharma'],color="black",linewidth=2,label="Rohit")
plt.title("Rohit sharma Vs Virat Kohli")
plt.xlabel("season")
plt.ylabel("Runs Scored")
plt.legend()
plt.legend(loc="upper left")
plt.show()
```
