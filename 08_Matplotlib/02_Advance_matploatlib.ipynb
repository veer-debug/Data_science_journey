```python
plt.figure(figsize=(15,7))
plt.scatter(df1['avg'],df1['strike_rate'],s=(df1['runs']))

for i in range(df1.shape[0]):
    plt.text(df1['avg'].values[i],df1['strike_rate'].values[i],df1['batter'].values[i])
plt.show()
```

## Horigental and vertical Line

```python
plt.figure(figsize=(15,7))
plt.scatter(df1['avg'],df1['strike_rate'],s=(df1['runs'])*3)
plt.axhline(120,color='green')
plt.axhline(60,color='orange')
plt.axvline(20,color='red')
for i in range(df1.shape[0]):
    plt.text(df1['avg'].values[i],df1['strike_rate'].values[i],df1['batter'].values[i])
plt.show()
```
