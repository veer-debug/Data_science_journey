```python
def fetch_data(url1):
    url=in_url+url1
    d=requests.get(url)
    soup=BeautifulSoup(d.text,"html.parser")
    a=soup.find_all('p')
    b=soup.find_all('ol')
    data['Section'].append(soup.find('h1').string)
    print(soup.find('h1').string)
    offence=None
    punisment=None
    cognizance=None
    bail=None
    trial=None
    discription=None





    if len(a)>=6:
        offence=a[1].string
        punisment=a[2].string
        cognizance=a[3].string
        bail=a[4].string
        trial=a[5].string
        discription=a[0].string
        

    elif len(b)==5:
        offence=b[0].text
        punisment=b[1].text
        cognizance=b[2].text
        bail=b[3].text
        trial=b[4].text
        discription=a[0].string
    elif len(b)>5:
        discription=b[0].text
        offence=b[1].text
        punisment=b[2].text
        cognizance=b[3].text
        bail=b[4].text
        trial=b[5].text
        
    data['Offence'].append(offence)
    data['Discription'].append(discription)
    data['Bail'].append(bail)
    data['Punishment'].append(punisment)
    data['Triable_By'].append(trial)
    data['Congizance'].append(cognizance)
       
```

```python
for i in l1:
    fetch_data(i)
    
```
