```python
import requests
from bs4 import BeautifulSoup
```

```python
def fetch(url,path):
    r=requests.get(url)
    with open(path,'w') as f:
        f.write(r.text)
```

```python
url='https://www.scribd.com/doc/36540419/HR-EMail-IDs-of-Top-500-Indian-Companies'
```

```python
r=requests.get(url)
```

```python
fetch(url,"data/test1.html")
```

```python
with open('data/test1.html','r') as f:
        html_doc=f.read()
```

```python
soup=BeautifulSoup(html_doc,"html.parser")
```
